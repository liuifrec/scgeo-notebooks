# 0x_gse280305_annotation.ipynb
## Coarse Cell-State Annotation and Radiation Program Scoring

---

## 0. Notebook purpose and scope

**Purpose**

This notebook establishes a biologically interpretable annotation layer for GSE280305 before any advanced geometric or dynamical analysis.

**Key goals**
- Assign coarse cell-state labels to clusters
- Quantify radiation-associated transcriptional programs
- Produce a stable annotated AnnData object for downstream ScGeo and Census analysis

**Out of scope**
- RNA velocity
- CellRank
- ScGeo metrics
- Census mapping

This notebook focuses purely on biological interpretation.

---

## 1. Load the consolidated dataset

**What is loaded**
- Consolidated AnnData object with:
  - cells × genes matrix
  - PCA / UMAP embeddings
  - clustering (`leiden_raw`)
  - timepoint metadata

**Rationale**
- Annotation should be performed on a stable artifact
- Re-running preprocessing is unnecessary and unsafe for memory

## 3. Marker-based cluster annotation

### 3.1 Rationale for coarse annotation

**Why coarse labels**
- Radiation disrupts transcriptional programs
- Fine-grained lineage annotation is unstable under stress
- Coarse labels are robust and interpretable

Target resolution: **broad cell-state categories**

---

### 3.2 Identify marker genes per cluster

**Method**
- Rank genes by `leiden_raw`

**Goal**
- Identify dominant transcriptional programs per cluster

Expected themes:
- HSC / LSK-like
- Cycling / proliferative
- Myeloid-primed
- Erythroid / megakaryocyte-primed
- Stress / DNA damage / interferon response

---

### 3.3 Assign coarse annotations

**Action**
- Manually map `leiden_raw → anno_coarse`

**Output**
- New column: `obs["anno_coarse"]`

**Interpretation**
- Biological state labels replace cluster IDs
- Serves as primary stratifier downstream

---

## 4. Radiation-associated transcriptional programs

### 4.1 Motivation

Radiation effects can appear as:
- changes in state composition
- activation of stress programs within states

Program scores disentangle these effects.

---

### 4.2 DNA damage / p53 response score

**Output**
- `obs["score_ddr"]`

**Interpretation**
- Indicates radiation-induced DNA damage signaling

---

### 4.3 Interferon / ISG score

**Output**
- `obs["score_ifn"]`

**Interpretation**
- Captures immune/stress response to irradiation

---

### 4.4 Optional: general stress / AP-1 score

**Purpose**
- Capture immediate early stress responses
- Distinguish transient stress from long-term identity changes
