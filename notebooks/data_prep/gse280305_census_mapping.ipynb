# 0xx_gse280305_census_mapping.ipynb
## Census-Based Reference Mapping for Radiation-Exposed HSCs

---

## 0. Notebook purpose and scope

**Purpose**
- Apply cellxgene Census mapping as an external reference
- Validate and contextualize coarse annotations

Census is used as **external evidence**, not ground truth.

---

## 1. Load annotated dataset

**Input**
- `GSE280305_annotated_coarse.h5ad`

**Rationale**
- Prior annotation is required to interpret mapping confidence and OOD behavior

---

## 2. Representation choice for mapping

**Primary representation**
- PCA (`X_pca_raw`)

**Rationale**
- Linear, stable, reproducible
- Avoids UMAP distortions

---

## 3. Census reference selection

**Reference**
- Mouse bone marrow / hematopoiesis atlas

**Principle**
- Broad biological match preferred over specificity

---

## 4. Census mapping execution

**Method**
- `map_query_to_ref_pool_census`

**Outputs**
- Predicted reference labels
- Confidence, entropy, and margin metrics

---

## 5. Mapping quality assessment

### 5.1 UMAP colored by Census labels

**Interpretation**
- Agreement with coarse annotations
- Detection of unexpected mappings

---

### 5.2 Mapping confidence analysis

**Interpretation**
- Low confidence often indicates radiation-specific or transitional states

---

### 5.3 Agreement with coarse annotation

**Analysis**
- Cross-tabulation: `anno_coarse × census_label`

**Interpretation**
- Agreement validates annotation
- Systematic disagreement reveals limitations or novelty

---

## 6. Identification of OOD or radiation-specific cells

**Criteria**
- Low confidence
- High entropy
- High DDR / IFN scores

**Interpretation**
- Likely radiation-induced states absent from reference atlases

---

## 7. Save Census-augmented dataset

**Output**
- `GSE280305_annotated_census.h5ad`

---

## 8. Summary and handoff to ScGeo

**We now have**
- Coarse biological annotation
- External reference validation
- Stress-aware context

**Next step**
- Geometry-based ScGeo analysis stratified by biological state
